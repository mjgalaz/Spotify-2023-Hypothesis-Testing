**TEST DE WILCOXON**

Correlación entre listas de streams para categoría Alto y Bajo


*   bpm
*   danceability
*   valence
*   energy
*   acousticness
*   instrumentalness
*   liveness
*   speechiness













```python
import pandas as pd
from scipy.stats import mannwhitneyu

# Función para imprimir en negrita
def print_bold(text):
    print(f"\033[1m{text}\033[0m")

# Función para imprimir en rojo y negrita
def print_bold_red(text):
    print(f"\033[1;31m{text}\033[0m")

# Almacenar streams en listas por categoría
streams_alto_bpm = []
streams_bajo_bpm = []

# Iterar sobre las filas del DataFrame
for index, row in df.iterrows():
    streams = row['streams_limpio']
    # Usar comparación correcta para 'cat_fin_bpm'
    if row['cat_fin_bpm'] == 'Alto':
        streams_alto_bpm.append(streams)
    elif row['cat_fin_bpm'] == 'Bajo':
        streams_bajo_bpm.append(streams)

# Aplicar el test de Wilcoxon (Mann-Whitney U test) a cada característica
statistic, p_value = mannwhitneyu(streams_alto_bpm, streams_bajo_bpm, alternative='two-sided')

# Definir nivel de significancia
alpha =0.05

# Mostrar los resultados del test para cada característica
bold_text = "Característica de la música: 'bpm'"
print_bold(bold_text)
if p_value < alpha:
    print_bold_red("Se rechaza la hipótesis nula: Hay una diferencia significativa entre las categorías 'alto' y 'bajo'")
else:
    print("No se puede rechazar la hipótesis nula: No hay una diferencia significativa entre las categorías 'alto' y 'bajo'")
# Imprimir los resultados
print(f"Estadística  U: {statistic}")
print(f"Valor p: {p_value}")

# danceability
streams_alto_danceability = []
streams_bajo_danceability = []
for index, row in df.iterrows():
    streams = row['streams_limpio']
    if row['cat_fin_danceability'] == 'Alto':
        streams_alto_danceability.append(streams)
    elif row['cat_fin_danceability'] == 'Bajo':
        streams_bajo_danceability.append(streams)
statistic, p_value = mannwhitneyu(streams_alto_danceability, streams_bajo_danceability, alternative='two-sided')
alpha = 0.05
bold_text = "Característica de la música: 'danceability'"
print_bold(bold_text)
if p_value < alpha:
    print_bold_red("Se rechaza la hipótesis nula: Hay una diferencia significativa entre las categorías 'alto' y 'bajo'")
else:
    print("No se puede rechazar la hipótesis nula: No hay una diferencia significativa entre las categorías 'alto' y 'bajo'")
print(f"Estadística U: {statistic}")
print(f"Valor p: {p_value}")

# valence
streams_alto_valence = []
streams_bajo_valence = []
for index, row in df.iterrows():
    streams = row['streams_limpio']
    if row['cat_fin_valence'] == 'Alto':
        streams_alto_valence.append(streams)
    elif row['cat_fin_valence'] == 'Bajo':
        streams_bajo_valence.append(streams)
statistic, p_value = mannwhitneyu(streams_alto_valence, streams_bajo_valence, alternative='two-sided')
alpha = 0.05
bold_text = "Característica de la música: 'valence'"
print_bold(bold_text)
if p_value < alpha:
    print_bold_red("Se rechaza la hipótesis nula: Hay una diferencia significativa entre las categorías 'alto' y 'bajo'")
else:
    print("No se puede rechazar la hipótesis nula: No hay una diferencia significativa entre las categorías 'alto' y 'bajo'")
print(f"Estadística U: {statistic}")
print(f"Valor p: {p_value}")

# energy
streams_alto_energy = []
streams_bajo_energy = []
for index, row in df.iterrows():
    streams = row['streams_limpio']
    if row['cat_fin_energy'] == 'Alto':
        streams_alto_energy.append(streams)
    elif row['cat_fin_energy'] == 'Bajo':
        streams_bajo_energy.append(streams)
statistic, p_value = mannwhitneyu(streams_alto_energy, streams_bajo_energy, alternative='two-sided')
alpha = 0.05
bold_text = "Característica de la música: 'energy'"
print_bold(bold_text)
if p_value < alpha:
    print_bold_red("Se rechaza la hipótesis nula: Hay una diferencia significativa entre las categorías 'alto' y 'bajo'")
else:
    print("No se puede rechazar la hipótesis nula: No hay una diferencia significativa entre las categorías 'alto' y 'bajo'")
print(f"Estadística U: {statistic}")
print(f"Valor p: {p_value}")

# acousticness
streams_alto_acousticness = []
streams_bajo_acousticness = []
for index, row in df.iterrows():
    streams = row['streams_limpio']
    if row['cat_fin_acousticness'] == 'Alto':
        streams_alto_acousticness.append(streams)
    elif row['cat_fin_acousticness'] == 'Bajo':
        streams_bajo_acousticness.append(streams)
statistic, p_value = mannwhitneyu(streams_alto_acousticness, streams_bajo_acousticness, alternative='two-sided')
alpha = 0.05
bold_text = "Característica de la música: 'acousticness'"
print_bold(bold_text)
if p_value < alpha:
    print_bold_red("Se rechaza la hipótesis nula: Hay una diferencia significativa entre las categorías 'alto' y 'bajo'")
else:
    print("No se puede rechazar la hipótesis nula: No hay una diferencia significativa entre las categorías 'alto' y 'bajo'")
print(f"Estadística U: {statistic}")
print(f"Valor p: {p_value}")

# instrumentalness
streams_alto_instrumentalness = []
streams_bajo_instrumentalness = []
for index, row in df.iterrows():
    streams = row['streams_limpio']
    if row['cat_fin_instrumentalness'] == 'Alto':
        streams_alto_instrumentalness.append(streams)
    elif row['cat_fin_instrumentalness'] == 'Bajo':
        streams_bajo_instrumentalness.append(streams)
statistic, p_value = mannwhitneyu(streams_alto_instrumentalness, streams_bajo_instrumentalness, alternative='two-sided')
alpha = 0.05
bold_text = "Característica de la música: 'instrumentalness'"
print_bold(bold_text)
if p_value < alpha:
    print_bold_red("Se rechaza la hipótesis nula: Hay una diferencia significativa entre las categorías 'alto' y 'bajo'")
else:
    print("No se puede rechazar la hipótesis nula: No hay una diferencia significativa entre las categorías 'alto' y 'bajo'")
print(f"Estadística U: {statistic}")
print(f"Valor p: {p_value}")

# liveness
streams_alto_liveness = []
streams_bajo_liveness = []
for index, row in df.iterrows():
    streams = row['streams_limpio']
    if row['cat_fin_liveness'] == 'Alto':
        streams_alto_liveness.append(streams)
    elif row['cat_fin_liveness'] == 'Bajo':
        streams_bajo_liveness.append(streams)
statistic, p_value = mannwhitneyu(streams_alto_liveness, streams_bajo_liveness, alternative='two-sided')
alpha = 0.05
bold_text = "Característica de la música: 'liveness'"
print_bold(bold_text)
if p_value < alpha:
    print_bold_red("Se rechaza la hipótesis nula: Hay una diferencia significativa entre las categorías 'alto' y 'bajo'")
else:
    print("No se puede rechazar la hipótesis nula: No hay una diferencia significativa entre las categorías 'alto' y 'bajo'")
print(f"Estadística U: {statistic}")
print(f"Valor p: {p_value}")

# speechiness
streams_alto_speechiness = []
streams_bajo_speechiness = []

# Iterar sobre las filas del DataFrame
for index, row in df.iterrows():
    streams = row['streams_limpio']
    if row['cat_fin_speechiness'] == 'Alto':
        streams_alto_speechiness.append(streams)
    elif row['cat_fin_speechiness'] == 'Bajo':
        streams_bajo_speechiness.append(streams)

# Verificar y eliminar valores NaN
streams_alto_speechiness = [x for x in streams_alto_speechiness if pd.notna(x)]
streams_bajo_speechiness = [x for x in streams_bajo_speechiness if pd.notna(x)]

# Imprimir siempre el bold_text
bold_text = "Característica de la música: 'speechiness'"
print_bold(bold_text)

# Verificar que ambos conjuntos de datos tengan algún tamaño
if len(streams_alto_speechiness) > 0 and len(streams_bajo_speechiness) > 0:
    try:
        statistic, p_value = mannwhitneyu(streams_alto_speechiness, streams_bajo_speechiness, alternative='two-sided')
        alpha = 0.05
        if p_value < alpha:
            print_bold_red("Se rechaza la hipótesis nula: Hay una diferencia significativa entre las categorías 'alto' y 'bajo'")
        else:
            print("No se puede rechazar la hipótesis nula: No hay una diferencia significativa entre las categorías 'alto' y 'bajo'")
        print(f"Estadística U: {statistic}")
        print(f"Valor p: {p_value}")
    except Exception as e:
        print_bold_red(f"Error: {e}")
else:
    print_bold_red("No hay datos válidos para realizar la prueba con esta variable")
```

**WILCOXON PARA COMPROBAR H12**
**"Las canciones mas populares en spotify tienen mayor número de streams"**

Calculo para  fecha de lanzamiento total de años, año 2022 y año 2023


*   streams_limpio
*   cat_fin_in_spotify_charts



```python
import pandas as pd
from scipy.stats import mannwhitneyu

# Función para imprimir en negrita
def print_bold(text):
    print(f"\033[1m{text}\033[0m")

# Función para imprimir en rojo y negrita
def print_bold_red(text):
    print(f"\033[1;31m{text}\033[0m")

# Leer el DataFrame (asegúrate de cargar tus datos correctamente)
# df = pd.read_csv("tu_archivo.csv")

# Asumiendo que ya tienes el DataFrame df con las columnas necesarias

# Convertir la columna 'fecha_completa_lanzamiento' a tipo datetime
df['fecha_completa_lanzamiento'] = pd.to_datetime(df['fecha_completa_lanzamiento'])

# Filtrar valores diferentes de cero en 'in_spotify_charts'
df_filtered = df[df['in_spotify_charts'] != 0].copy()

# Calcular los cuartiles para el conjunto completo de datos
cuartiles_total = df_filtered['in_spotify_charts'].quantile([0.25, 0.75])

# Función para asignar etiquetas 'Alto' o 'Bajo' según los cuartiles
def etiquetar_cat_in_spotify_charts(valor):
    if valor <= cuartiles_total[0.25]:
        return 'Alto'
    else:
        return 'Bajo'

# Crear una nueva columna 'cat_fin_in_spotify_charts' usando la función de etiquetado
df_filtered['cat_fin_in_spotify_charts'] = df_filtered['in_spotify_charts'].apply(etiquetar_cat_in_spotify_charts)

# Almacenar streams en listas por categoría
streams_alto_in_spotify_charts = df_filtered[df_filtered['cat_fin_in_spotify_charts'] == 'Alto']['streams_limpio'].tolist()
streams_bajo_in_spotify_charts = df_filtered[df_filtered['cat_fin_in_spotify_charts'] == 'Bajo']['streams_limpio'].tolist()

# Aplicar el test de Wilcoxon (Mann-Whitney U test) a cada característica
try:
    statistic, p_value = mannwhitneyu(streams_alto_in_spotify_charts, streams_bajo_in_spotify_charts, alternative='two-sided')
    alpha = 0.05
    bold_text = "H12 TOTAL"
    print_bold(bold_text)
    if p_value < alpha:
        print_bold_red("Se rechaza la hipótesis nula: Hay una diferencia significativa entre las categorías 'Alto' y 'Bajo'")
    else:
        print("No se puede rechazar la hipótesis nula: No hay una diferencia significativa entre las categorías 'Alto' y 'Bajo'")
    print(f"Estadística U: {statistic}")
    print(f"Valor p: {p_value}")
except Exception as e:
    print_bold_red(f"Error: {e}")

# Filtrar datos para el año 2022 y asignar a una nueva variable usando .loc
df_2022 = df.loc[df['fecha_completa_lanzamiento'].dt.year == 2022].copy()

# Verificar si hay datos para el año 2022
if not df_2022.empty:
    # Filtrar valores diferentes de cero en 'in_spotify_charts'
    df_2022_filtered = df_2022[df_2022['in_spotify_charts'] != 0].copy()

    # Calcular los cuartiles para el conjunto de datos del año 2022
    cuartiles_2022 = df_2022_filtered['in_spotify_charts'].quantile([0.25, 0.75])

    # Función para asignar etiquetas 'Alto' o 'Bajo' según los cuartiles del año 2022
    def etiquetar_cat_in_spotify_charts(valor):
        if valor <= cuartiles_2022[0.25]:
            return 'Alto'
        else:
            return 'Bajo'

    # Crear una nueva columna 'cat_fin_in_spotify_charts' usando la función de etiquetado
    df_2022_filtered['cat_fin_in_spotify_charts'] = df_2022_filtered['in_spotify_charts'].apply(etiquetar_cat_in_spotify_charts)

    # Almacenar streams en listas por categoría
    streams_alto_in_spotify_charts = df_2022_filtered[df_2022_filtered['cat_fin_in_spotify_charts'] == 'Alto']['streams_limpio'].tolist()
    streams_bajo_in_spotify_charts = df_2022_filtered[df_2022_filtered['cat_fin_in_spotify_charts'] == 'Bajo']['streams_limpio'].tolist()

    # Aplicar el test de Wilcoxon (Mann-Whitney U test) a cada característica
    try:
        statistic, p_value = mannwhitneyu(streams_alto_in_spotify_charts, streams_bajo_in_spotify_charts, alternative='two-sided')
        alpha = 0.05
        bold_text = "H12 AÑO 2022"
        print_bold(bold_text)
        if p_value < alpha:
            print_bold_red("Se rechaza la hipótesis nula: Hay una diferencia significativa entre las categorías 'Alto' y 'Bajo'")
        else:
            print("No se puede rechazar la hipótesis nula: No hay una diferencia significativa entre las categorías 'Alto' y 'Bajo'")
        print(f"Estadística U: {statistic}")
        print(f"Valor p: {p_value}")
    except Exception as e:
        print_bold_red(f"Error: {e}")
else:
    print_bold_red("No hay datos válidos para realizar la prueba con la variable in_spotify_charts en el año 2022")

# Filtrar datos para el año 2023 y asignar a una nueva variable usando .loc
df_2023 = df.loc[df['fecha_completa_lanzamiento'].dt.year == 2023].copy()

# Verificar si hay datos para el año 2023
if not df_2023.empty:
    # Filtrar valores diferentes de cero en 'in_spotify_charts'
    df_2023_filtered = df_2023[df_2023['in_spotify_charts'] != 0].copy()

    # Calcular los cuartiles para el conjunto de datos del año 2023
    cuartiles_2023 = df_2023_filtered['in_spotify_charts'].quantile([0.25, 0.75])

    # Función para asignar etiquetas 'Alto' o 'Bajo' según los cuartiles del año 2023
    def etiquetar_cat_in_spotify_charts(valor):
        if valor <= cuartiles_2023[0.25]:
            return 'Alto'
        else:
            return 'Bajo'

    # Crear una nueva columna 'cat_fin_in_spotify_charts' usando la función de etiquetado
    df_2023_filtered['cat_fin_in_spotify_charts'] = df_2023_filtered['in_spotify_charts'].apply(etiquetar_cat_in_spotify_charts)

    # Almacenar streams en listas por categoría
    streams_alto_in_spotify_charts = df_2023_filtered[df_2023_filtered['cat_fin_in_spotify_charts'] == 'Alto']['streams_limpio'].tolist()
    streams_bajo_in_spotify_charts = df_2023_filtered[df_2023_filtered['cat_fin_in_spotify_charts'] == 'Bajo']['streams_limpio'].tolist()

    # Aplicar el test de Wilcoxon (Mann-Whitney U test) a cada característica
    try:
        statistic, p_value = mannwhitneyu(streams_alto_in_spotify_charts, streams_bajo_in_spotify_charts, alternative='two-sided')
        alpha = 0.05
        bold_text = "H12 AÑO 2023"
        print_bold(bold_text)
        if p_value < alpha:
            print_bold_red("Se rechaza la hipótesis nula: Hay una diferencia significativa entre las categorías 'Alto' y 'Bajo'")
        else:
            print("No se puede rechazar la hipótesis nula: No hay una diferencia significativa entre las categorías 'Alto' y 'Bajo'")
        print(f"Estadística U: {statistic}")
        print(f"Valor p: {p_value}")
    except Exception as e:
        print_bold_red(f"Error: {e}")
else:
    print_bold_red("No hay datos válidos para realizar la prueba con la variable in_spotify_charts en el año 2023")
```
